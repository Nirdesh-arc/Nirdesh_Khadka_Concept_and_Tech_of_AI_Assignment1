```python
#Task3. Outlier Detection

#IQR calculation for HDI
hdi_q1, hdi_q3 = south_asia_df["hdi"].quantile([0.25, 0.75])
hdi_iqr = hdi_q3 - hdi_q1

#IQR calculation for GNI per Capita
gni_q1, gni_q3 = south_asia_df["gross_inc_percap"].quantile([0.25, 0.75])
gni_iqr = gni_q3 - gni_q1

#Outlier condition
hdi_outlier = (
    (south_asia_df["hdi"] < hdi_q1 - 1.5 * hdi_iqr) |
    (south_asia_df["hdi"] > hdi_q3 + 1.5 * hdi_iqr)
)

gni_outlier = (
    (south_asia_df["gross_inc_percap"] < gni_q1 - 1.5 * gni_iqr) |
    (south_asia_df["gross_inc_percap"] > gni_q3 + 1.5 * gni_iqr)
)

south_asia_df["is_outlier"] = hdi_outlier | gni_outlier

#Split dataset
normal_points = south_asia_df.loc[~south_asia_df["is_outlier"]]
outlier_points = south_asia_df.loc[south_asia_df["is_outlier"]]

#Show detected outliers
outlier_points[["country", "hdi", "gross_inc_percap"]]
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=south_asia_df,
    x='gross_inc_percap',
    y='hdi',
    hue='is_outlier',
    palette={True: 'red', False: 'blue'}
)

plt.title("GNI per Capita vs HDI (South Asia)")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend(title="Outlier")
plt.show()
```

```python
#Task7. Outlier Detection (1.5  IQR Rule)
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)
```

```python
#Identify Outliers
for df_region in [south_asia_2020_22, middle_east_2020_22]:
    df_region['Outlier'] = (
        detect_outliers(df_region['hdi']) |
        detect_outliers(df_region['gross_inc_percap'])
    )

print("South Asia Outliers:")
print(south_asia_2020_22[south_asia_2020_22['Outlier']][
    ['country', 'hdi', 'gross_inc_percap']
])

print("\nMiddle East Outliers:")
print(middle_east_2020_22[middle_east_2020_22['Outlier']][
    ['country', 'hdi', 'gross_inc_percap']
])
```
